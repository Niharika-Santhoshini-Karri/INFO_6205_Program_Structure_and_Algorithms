1) Solve the following recurrence using the Master Theorem:

T(n) = 4T(n/2) + n^2 log n

Solution:

Applying the Master Theorem to the given recurrence, 

a = 4, b = 2, and f(n) = n^2 log n.

Comparing f(n) to n^log_b(a),

n^log_b(a) = n^log_2(4) = n^2

 Compare f(n) to n^2 log n. Since f(n) is not polynomially smaller or larger than n^2 log n, the Master Theorem doesn't directly apply.

use the regularity condition to check the case:

af(n/b) ≤ kf(n) for some constant k < 1, and all sufficiently large n.

4(f(n/2)) ≤ kf(n)

f(n/2) ≤ (k/4)f(n)

n^2/2 log(n/2) ≤ (k/4)n^2 log n

n^2 log(n/2) ≤ (k/4)n^2 log n

n^2 log n - n^2 ≤ (k/4)n^2 log n

n^2 log n ≤ (k/4)n^2 log n + n^2

Dividing both sides by n^2 log n:

1 ≤ (k/4) + 1/n

As n approaches infinity, the right-hand side approaches (k/4), so k must be less than 4.

Now, the three cases are:

If k < 1/4 (i.e., k = 0), then T(n) = Theta(n^2 log n).

If k = 1/4, then T(n) = Theta(n^2 log n * log n).

If k > 1/4, then T(n) = Theta(n^2 log n).

4) Given an unsorted array of integers, find the length of the longest increasing subsequence. An increasing subsequence is a subsequence in which elements are in strictly increasing order.

Write a polynomial-time algorithm to find the length of the longest increasing subsequence

```python
def length_of_lis(nums):
    if not nums:
        return 0

    n = len(nums)
    lis = [1] * n

    for i in range(1, n):
        for j in range(i):
            if nums[i] > nums[j]:
                lis[i] = max(lis[i], lis[j] + 1)

    return max(lis)

nums = [10, 22, 9, 33, 21, 50, 41, 60, 80]
result = length_of_lis(nums)
print("Length of Longest Increasing Subsequence:", result)
```

6) You are given an array of n integers, where each integer represents the price of a stock on a given day. Your goal is to maximize your profit by buying and selling stocks, but with the following constraints:

You can only perform at most k transactions, buy and sell a stock is counted as one transaction.
You cannot buy another stock before selling the one you already have.
Design a polynomial-time algorithm that finds the maximum profit you can achieve under these constraints.

Solution:

This problem can be solved using dynamic programming. use a 3D array dp to keep track of the maximum profit. Here's a step-by-step solution:

Initialize a 3D array dp, where dp[i][j][k] represents the maximum profit on day i, with j transactions, and whether or not we hold a stock (k = 0 for not holding, k = 1 for holding).

Initialize dp[i][0][0] and dp[i][0][1] for all i to 0, as there's no profit with zero transactions.

For each day i from 1 to n:

For each transaction j from 1 to k:
Calculate dp[i][j][0] (not holding):
a. dp[i-1][j][0] (no change, didn't buy/sell)
b. dp[i-1][j][1] + prices[i] (sell on this day, adding profit)
Calculate dp[i][j][1] (holding):
a. dp[i-1][j][1] (no change, still holding)
b. dp[i-1][j-1][0] - prices[i] (buy on this day)
The answer is the maximum of dp[n][j][0] for all j (maximum profit with j transactions and not holding a stock).

The time complexity of this dynamic programming solution is O(n * k), which is polynomial time. This algorithm efficiently finds the maximum profit while respecting the transaction limit and the constraint of not buying a new stock before selling the one you already have.

8) You are given a list of cities and the distances between each pair of cities. Your task is to find the shortest possible route that visits each city exactly once and returns to the original city.

Write a polynomial-time algorithm to solve the Traveling Salesman Problem (TSP)

```python
import itertools

def tsp(graph, start=0):
    n = len(graph)
    cities = range(n)
    all_tours = itertools.permutations(cities)
    min_tour = None
    min_dist = float("inf")

    for tour in all_tours:
        if tour[0] != start:
            continue

        tour_len = 0
        prev_city = start

        for city in tour:
            tour_len += graph[prev_city][city]
            prev_city = city

        tour_len += graph[prev_city][start]

        if tour_len < min_dist:
            min_dist = tour_len
            min_tour = tour

    return min_dist, min_tour


graph = [
    [0, 29, 20, 21],
    [29, 0, 15, 18],
    [20, 15, 0, 12],
    [21, 18, 12, 0]
]

min_distance, min_tour = tsp(graph)
print("Minimum Distance:", min_distance)
print("Optimal Tour:", min_tour)
```

9)You are given an array of N positive integers. Find the length of the longest increasing subsequence in the array. However, you are not allowed to use dynamic programming or recursion. Design a polynomial-time algorithm to solve this problem without these common techniques.

Solution :

You can solve this problem using a modified binary search approach. The algorithm follows these steps:

Initialize an empty array tails of length N. The tails array will store the smallest tail element of all increasing subsequences found so far.

Iterate through the input array, and for each element x:

If x is greater than the last element in tails, append x to tails. Otherwise, find the smallest element in tails that is greater than or equal to x and replace it with x using binary search. The length of the tails array is the length of the longest increasing subsequence.

This algorithm has a time complexity of O(N log N) and does not rely on dynamic programming or recursion.



10) You are given a set of n intervals on the real number line. Your task is to find the maximum number of disjoint intervals that can be selected. Design a polynomial-time algorithm to solve this problem.

Solution:

To find the maximum number of disjoint intervals, you can use the following algorithm:

Sort the intervals based on their endpoint in non-decreasing order.

Initialize a variable count to 1 and a variable end to the endpoint of the first interval.

Iterate through the sorted intervals starting from the second interval:

If the current interval's start is greater than or equal to end, it can be selected. Update end to the endpoint of the current interval and increment count by 1. The value of count represents the maximum number of disjoint intervals that can be selected.

This algorithm has a time complexity of O(n log n) due to the sorting step and runs in polynomial time.
